```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```python
alice_fd = nltk.FreqDist(alice)
```

```python
alice_fd
```

```python
alice_fd["Alice"]
```

```python
alice_fd['rabbit']
```

```python
alice_fd['Rabbit']
```

```python
alice_fd.most_common(15)
```

```python
# Get the list of words appearing only once in the text
alice_fd.hapaxes()
```
